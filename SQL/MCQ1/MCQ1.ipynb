```python
%%sql
SELECT title, popularity,
rank() OVER (ORDER BY popularity)
FROM Movies
WHERE original_language="en" and release_date > "2010-01-01" and release_date < "2016-01-01" and title like '%love%' and budget > 10000000;
```
